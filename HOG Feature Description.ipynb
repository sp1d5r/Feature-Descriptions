# Histogram of Oriented Gradients 
This is another feature detection algorithm. It's similar to SIFT because it uses the gradient orientation for feature description but it has a key difference:
* HOG describes features for a large region. 

Okay so how does it work? 
* For starters look at each pixel, and calculate the gradient orientation and magnitude of it.
    * you calculate the derivative of the image in the x direction using the filter [1,0,-1] and then for y using [1,0,-1]^T
    * Then you calculate the magnitude of that using sqrt(d/dx^2 + d/dy^2) and finally you get the orientation of that using arctan(d/dx / d/dy)
* Okay now divide the image up into squares for example you can split it up into blocks of 20 x 20 pixels. 
* Then for each block you decide on a histogram of gradients, you can use 8 bins. [0, 45, 90, ..., 315] and add in the magnitude for each one.
* Then you represent this in the image

```python
#Â Import all the libraries 
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
```

```python
img = imread('IMG_0030.jpeg')
plt.axis("off")
plt.imshow(img)
print(img.shape)
```

```python
fd, hog_image = hog(img, orientations=8, pixels_per_cell=(20, 20),
                	cells_per_block=(2, 2), visualize=True, multichannel=True)
plt.axis("off")
plt.imshow(hog_image, cmap="gray")
```

Some important things about HoG compared to the SIFT is that HoG is not rotationally invariant or scale invariant. But with normalisation it becomes invariant to illumination. Whereas SIFT is invariant to rotation, scale and intensity since it uses histograms
